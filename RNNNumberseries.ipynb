```python
data=[6,12,18,24,30,36,42,48,54,60,66,72,78,84,90,96,102,108,114,120,126,132,138,144,150,156,162,168,174,180,186,192,198,204,210,216,222,228,234,240]
```

```python
import numpy as np

def splitdata(data,n):
  x=[]
  y=[]

  for i in range (len(data)):
    last=i+n

    if last > len(data)-1:
      break

    seq_x,seq_y=data[i:last],data[last]

    x.append(seq_x)
    y.append(seq_y)

  x=np.array(x)
  y=np.array(y)

  return x,y
```

```python
split=4
x,y=splitdata(data,split)

print(x)
```

```python
n_features=1
print(x.shape)
x=x.reshape(x.shape[0], x.shape[1], n_features)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import *
from keras.models import *
```

```python
model=Sequential()
model.add(LSTM(50,activation='relu',input_shape=(split,n_features)))
model.add(Dense(1))
model.layers
model.compile(optimizer="Adam",loss='mse',metrics='accuracy')
model.fit(x,y,epochs=300,verbose=1)
```

```python
test=np.array([60,66,72,78])
test=test.reshape(1,split,n_features)
res = model.predict(test)
res
```
